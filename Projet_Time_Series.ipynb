```python
## Question 1

import pandas as pd
import matplotlib.pyplot as plt

dpaes = pd.read_excel("/Users/baba/Desktop/dpae-mensuelles-france-entiere.xlsx",names= ['Année' ,'Trimestre','Dernier jour du mois','Durée de contrat','Nature de contrat','DPAE (brut)','DPAE (cvs)'])

# Convertir la colonne 'Dernier jour du mois' en type datetime
dpaes['Dernier jour du mois'] = pd.to_datetime(dpaes['Dernier jour du mois'])

cdd = dpaes[dpaes["Nature de contrat"] == "CDD de plus d'un mois"]
cdi = dpaes[dpaes["Nature de contrat"] == "CDI"]

cdd = cdd.sort_values(by='Dernier jour du mois')
cdi = cdi.sort_values(by='Dernier jour du mois')




## Pour les CDD



# Tracer la variation de DPAE (brut) au fur et à mesure du temps pour les CDD
plt.figure(figsize=(12, 6))
plt.plot(cdd['Dernier jour du mois'], cdd['DPAE (brut)'])
plt.xlabel('Date')
plt.ylabel('DPAE (brut)')
plt.title('Variation de DPAE (brut) pour les CDD au fur et à mesure du temps')
plt.show()

plt.figure()

# Calculer la moyenne et l'écart type de DPAE (brut) par année 
mean_by_year = cdd.groupby('Année')['DPAE (brut)'].mean()
std_by_year = cdd.groupby('Année')['DPAE (brut)'].std()
median_by_year = cdd.groupby('Année')['DPAE (brut)'].median()

mean_by_year.plot()
std_by_year.plot()
median_by_year.plot()
plt.legend(['Mean', 'Standard Deviation', 'Median'])
plt.title('Statistiques de DPAE (brut) pour les CDD au fur et à mesure des années')
plt.show()


# Calculer la moyenne et l'écart type de DPAE (brut) par trimestre
mean_by_trimester = cdd.groupby(['Année', 'Trimestre'])['DPAE (brut)'].mean()
std_by_trimester = cdd.groupby(['Année', 'Trimestre'])['DPAE (brut)'].std()
median_by_trimester = cdd.groupby(['Année','Trimestre'])['DPAE (brut)'].median()

plt.figure()
mean_by_trimester.plot()
std_by_trimester.plot()
median_by_trimester.plot()
plt.title('Statistiques de DPAE (brut) pour les CDD au fur et à mesure des années')
plt.legend(['Mean', 'Standard Deviation', 'Median'])
plt.show()

# Afficher les boîtes à moustaches par année
cdd.boxplot(column=['DPAE (brut)'], by='Année', figsize=(10, 6))
plt.title("Boîtes à moustaches pour les CDD par année")
plt.ylabel("DPAE (brut)")
plt.show()

# Afficher les boîtes à moustaches par trimestre
cdd.boxplot(column=['DPAE (brut)'], by='Trimestre', figsize=(10, 6))
plt.title("Boîtes à moustaches pour les CDD par trimestre")
plt.ylabel("DPAE (brut)")
plt.show()


```

```python
## Pour les CDI


# Tracer la variation de DPAE (brut) au fur et à mesure du temps pour les CDI
plt.figure(figsize=(12, 6))
plt.plot(cdi['Dernier jour du mois'], cdi['DPAE (brut)'], label ='CDI')
plt.xlabel('Date')
plt.ylabel('DPAE (brut)')
plt.title('Variation de DPAE (brut) pour les CDI au fur et à mesure du temps')
plt.show()



plt.figure()

# Calculer la moyenne et l'écart type de DPAE (brut) par année 
mean_by_year = cdi.groupby('Année')['DPAE (brut)'].mean()
std_by_year = cdi.groupby('Année')['DPAE (brut)'].std()
median_by_year = cdi.groupby('Année')['DPAE (brut)'].median()

mean_by_year.plot()
std_by_year.plot()
median_by_year.plot()
plt.legend(['Mean', 'Standard Deviation', 'Median'])
plt.title('Statistiques de DPAE (brut) pour les CDI au fur et à mesure des années')
plt.show()


# Calculer la moyenne et l'écart type de DPAE (brut) par trimestre
mean_by_trimester = cdi.groupby(['Année', 'Trimestre'])['DPAE (brut)'].mean()
std_by_trimester = cdi.groupby(['Année', 'Trimestre'])['DPAE (brut)'].std()
median_by_trimester = cdi.groupby(['Année','Trimestre'])['DPAE (brut)'].median()

plt.figure()
mean_by_trimester.plot()
std_by_trimester.plot()
median_by_trimester.plot()
plt.title('Statistiques de DPAE (brut) pour les CDD au fur et à mesure des trimestres')
plt.legend(['Mean', 'Standard Deviation', 'Median'])
plt.show()

# Afficher les boîtes à moustaches par année
cdi.boxplot(column=['DPAE (brut)'], by='Année', figsize=(10, 6))
plt.title("Boîtes à moustaches pour les CDI par année")
plt.ylabel("DPAE (brut)")
plt.show()

# Afficher les boîtes à moustaches par trimestre
cdi.boxplot(column=['DPAE (brut)'], by='Trimestre', figsize=(10, 6))
plt.title("Boîtes à moustaches pour les CDI par trimestre")
plt.ylabel("DPAE (brut)")
plt.show()


```

```python
## Pour le deuxième fichier 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Lire les données à partir des feuilles Excel
df_valeurs = pd.read_excel('/Users/baba/Desktop/Time Series/famille_EMPLOI-SALARIE-TRIM-NATIONAL_09032023.xlsx', sheet_name='valeurs_trimestrielles')
df_caract = pd.read_excel('/Users/baba/Desktop/Time Series/famille_EMPLOI-SALARIE-TRIM-NATIONAL_09032023.xlsx', sheet_name='caractéristiques')
df_codes = pd.read_excel('/Users/baba/Desktop/Time Series/famille_EMPLOI-SALARIE-TRIM-NATIONAL_09032023.xlsx', sheet_name='codes')

# Fusionner les données en utilisant la colonne 'idBank'
df_eessa = pd.merge(df_caract, df_valeurs, on='idBank')
df_eessa = df_eessa.sort_values(by='idBank')

# Renommer la colonne "Libellé" en "Secteur"
df_eessa.rename(columns={'Libellé': 'Secteur'}, inplace=True)

# Nettoyer les données en remplaçant les valeurs manquantes "(O)" par des NaN
df_eessa = df_eessa.replace('(O)', np.nan)

# Extraire les valeurs trimestrielles des données EESSA
start_date = '1970-T4'
end_date = '2022-T4'
df_essaa_values = df_eessa.loc[:, start_date:end_date]
df_essaa_values['Secteur'] = df_eessa['Secteur']
df_essaa_values = df_essaa_values.set_index('Secteur').T

# Calculer la somme de tous les secteurs à chaque instant
df_valsum_secteur = df_essaa_values.sum(axis=1)

# Convertir l'index en année
year_index = pd.to_datetime(df_valsum_secteur.index).year

# Afficher les résultats
plt.figure(figsize=(12, 6))
plt.plot(year_index, df_valsum_secteur.values, color='b')
plt.title('Somme des valeurs trimestrielles par secteur (1970-2022)')
plt.xlabel('Année')
plt.ylabel('Somme des valeurs')
plt.grid(True)
plt.show()



```

```python

## Pour les CDD

from statsmodels.tsa.seasonal import seasonal_decompose

fig = plt.figure(figsize=(12, 6)) # définit une figure de taille 12x6 pouces


result_cdd = seasonal_decompose(cdd['DPAE (brut)'], period=12)
# Afficher les composantes de la décomposition
plt.plot(cdd['Dernier jour du mois'],result_cdd.trend)
plt.plot(cdd['Dernier jour du mois'],result_cdd.seasonal)
plt.plot(cdd['Dernier jour du mois'],result_cdd.resid)
plt.title('Analyse de DPAE (brut) pour les CDD au fur et à mesure des trimestres')
plt.legend(['Trend', 'Seasonal', 'Resid'])
plt.show()


## Pour les CDI

fig = plt.figure(figsize=(12, 6)) # définit une figure de taille 12x6 pouces


result_cdi = seasonal_decompose(cdi['DPAE (brut)'], period=12)
# Afficher les composantes de la décomposition
plt.plot(cdi['Dernier jour du mois'],result_cdi.trend)
plt.plot(cdi['Dernier jour du mois'],result_cdi.seasonal)
plt.plot(cdi['Dernier jour du mois'],result_cdi.resid)
plt.title('Analyse de DPAE (brut) pour les CDI au fur et à mesure des trimestres')
plt.legend(['Trend', 'Seasonal', 'Resid'])
plt.show()



## Pour les secteurs eessa



    
fig = plt.figure(figsize=(12, 6)) # définit une figure de taille 12x6 pouces
result_sec1 = seasonal_decompose(df_essaa_values[Secteur1].values, period=12)
result_sec2 = seasonal_decompose(df_essaa_values[Secteur2].values, period=12)
result_sec3 = seasonal_decompose(df_essaa_values[Secteur3].values, period=12)



# Afficher les composantes de la décomposition

plt.plot(df_essaa_values[Secteur1].index,result_sec1.trend)
plt.plot(df_essaa_values[Secteur2].index,result_sec2.trend)
plt.plot(df_essaa_values[Secteur3].index,result_sec3.trend)
plt.title('trend')
plt.legend(['Secteur1','Secteur2', 'Secteur3'])
plt.show()


fig = plt.figure(figsize=(12, 6)) # définit une figure de taille 12x6 pouces
plt.plot(df_essaa_values[Secteur1].index,result_sec1.seasonal)
plt.plot(df_essaa_values[Secteur2].index,result_sec2.seasonal)
plt.plot(df_essaa_values[Secteur3].index,result_sec3.seasonal)
plt.title('seasonal')
plt.legend(['Secteur1','Secteur2', 'Secteur3'])

plt.show()


fig = plt.figure(figsize=(12, 6)) # définit une figure de taille 12x6 pouces
plt.plot(df_essaa_values[Secteur1].index,result_sec1.resid)
plt.plot(df_essaa_values[Secteur2].index,result_sec2.resid)
plt.plot(df_essaa_values[Secteur3].index,result_sec3.resid)
plt.title('resid')
plt.legend(['Secteur1','Secteur2', 'Secteur3'])
plt.show()


```

```python
# Question 3


## Pour le fichier EESSA

fig = plt.figure(figsize=(12, 6)) # définit une figure de taille 12x6 pouces

data_1 = df_essaa_values[Secteur1].values.tolist()
data_2 = df_essaa_values[Secteur2].values.tolist()
data_3 = df_essaa_values[Secteur3].values.tolist()




# Diviser l'ensemble de données en ensembles d'apprentissage et de test (fractionnement à 80 %)
train_size = int(len(data_cdd) * 0.75)
train_c1, test_1 = data_1[:train_size], data_1[train_size:]
train_c2, test_2 = data_2[:train_size], data_2[train_size:]
train_c3, test_3 = data_3[:train_size], data_3[train_size:]


from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt

fit1 = SimpleExpSmoothing(train_c1, initialization_method="heuristic").fit(
    smoothing_level=0.2, optimized=False)



fig = plt.figure(figsize=(12, 6)) # définit une figure de taille 12x6 pouces



from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt

fit2 = SimpleExpSmoothing(train_c2, initialization_method="heuristic").fit(
    smoothing_level=0.2, optimized=False)

fit3 = SimpleExpSmoothing(train_c3, initialization_method="heuristic").fit(
    smoothing_level=0.2, optimized=False)


plt.plot(df_essaa_values[Secteur1].index[:train_size],train_c1, marker="o", color="black", label='DPAE(brut)')
plt.plot(df_essaa_values[Secteur1].index[:train_size],fit1.fittedvalues, marker="o", color="blue", label='forecast')
plt.title('prédictions par lissage exponentiel sur les données secteur1')
plt.legend()
plt.show()

plt.plot(df_essaa_values[Secteur2].index[:train_size],train_c2, marker="o", color="black", label='DPAE(brut)')
plt.plot(df_essaa_values[Secteur2].index[:train_size],fit2.fittedvalues, marker="o", color="blue", label='forecast')
plt.title('prédictions par lissage exponentiel sur les données secteur2')

plt.legend()
plt.show()

plt.plot(df_essaa_values[Secteur3].index[:train_size],train_c3, marker="o", color="black", label='DPAE(brut)')
plt.plot(df_essaa_values[Secteur3].index[:train_size],fit3.fittedvalues, marker="o", color="blue", label='forecast')
plt.title('prédictions par lissage exponentiel sur les données secteur3')

plt.legend()
plt.show()

```

```python
# Question 4

from sklearn.metrics import mean_squared_error
# Calculer les prédictions
forecast_cdd = fit1.forecast(len(test_cdd))
forecast_cdi = fit2.forecast(len(test_cdi))

forecast_1 = fit1.forecast(len(test_1))
forecast_2 = fit2.forecast(len(test_2))
forecast_3 = fit3.forecast(len(test_3))



# Concaténer les valeurs réelles et les valeurs prédites
y_true_cdd = test_cdd
y_pred_cdd = forecast_cdd

y_true_cdi = test_cdi
y_pred_cdi = forecast_cdi

y_true_1 = test_1
y_pred_1 = forecast_1

y_true_2 = test_2
y_pred_2 = forecast_2

y_true_3 = test_3
y_pred_3 = forecast_3


# Calculer l'erreur MSE
rmse_cdd = mean_squared_error(y_true_cdd, y_pred_cdd, squared=False)
rmse_cdi = mean_squared_error(y_true_cdi, y_pred_cdi, squared=False)
rmse_1 = mean_squared_error(y_true_1, y_pred_1, squared=False)
rmse_2 = mean_squared_error(y_true_2, y_pred_2, squared=False)
rmse_3 = mean_squared_error(y_true_3, y_pred_3, squared=False)


# Afficher l'erreur MSE
print("Rooted Mean Squared Error (RMSE) CDD: ", rmse_cdd)
print("Rooted Mean Squared Error (RMSE) CDI: ", rmse_cdi)
print("Rooted Mean Squared Error (RMSE) secteur1: ", rmse_1)
print("Rooted Mean Squared Error (RMSE) secteur2: ", rmse_2)
print("Rooted Mean Squared Error (RMSE) secteur3: ", rmse_3)


```

```python
## Question 5
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm

# Entraînement du modèle ARIMA pour DPAE CDD
model_dpaecdd = ARIMA(train_cdd, order=(2, 0, 1)).fit()

# Entraînement du modèle ARIMA pour DPAE CDI
model_dpaecdi = ARIMA(train_cdi, order=(2, 0, 1)).fit()

# Entraînement du modèle ARIMA pour EESSA (secteur sélectionné)
model_eessa = sm.tsa.ARIMA(train_c1, order=(2,0, 1)).fit()
model_eessa2 = sm.tsa.ARIMA(train_c2, order=(2,0, 1)).fit()
model_eessa3 = sm.tsa.ARIMA(train_c3, order=(2,0, 1)).fit()



# Prévisions pour DPAE CDD
pred_dpaecdd = model_dpaecdd.predict(start=len(train_cdd), end=len(train_cdd)+len(test_cdd)-1)

# Prévisions pour DPAE CDI
pred_dpaecdi = model_dpaecdi.predict(start=len(train_cdi), end=len(train_cdi)+len(test_cdi)-1)

# Prévisions pour EESSA (secteur sélectionné)
pred_eessa = model_eessa.predict(start=len(train_c1), end=len(train_c1)+len(test_1)-1)
pred_eessa2 = model_eessa2.predict(start=len(train_c2), end=len(train_c2)+len(test_2)-1)
pred_eessa3 = model_eessa3.predict(start=len(train_c3), end=len(train_c3)+len(test_3)-1)


from sklearn.metrics import mean_absolute_error, mean_squared_error
from scipy.stats import pearsonr

# Évaluation des prévisions pour DPAE CDD
mae_cdd_arima = mean_absolute_error(test_cdd, pred_dpaecdd)
rmse_cdd_arima = mean_squared_error(test_cdd, pred_dpaecdd, squared=False)
print("RMSE pour les données DPAE (CDD) avec ARIMA:", rmse_cdd_arima)
# Évaluation des prévisions pour DPAE CDI
mae_cdi_arima = mean_absolute_error(test_cdi, pred_dpaecdd)
rmse_cdi_arima = mean_squared_error(test_cdi, pred_dpaecdd, squared=False)
print("RMSE pour les données DPAE (CDI) avec ARIMA:", rmse_cdi_arima)

# Évaluation des prévisions pour Secteur1
mae_1_arima = mean_absolute_error(test_1, pred_eessa)
rmse_1_arima = mean_squared_error(test_1, pred_eessa, squared=False)
print("RMSE pour les données EESSA avec ARIMA:", rmse_1_arima)

# Évaluation des prévisions pour Secteur2
mae_2_arima = mean_absolute_error(test_2, pred_eessa2)
rmse_2_arima = mean_squared_error(test_2, pred_eessa2, squared=False)
print("RMSE pour les données EESSA avec ARIMA:", rmse_2_arima)

# Évaluation des prévisions pour Secteur3
mae_3_arima = mean_absolute_error(test_3, pred_eessa3)
rmse_3_arima = mean_squared_error(test_3, pred_eessa3, squared=False)
print("RMSE pour les données EESSA avec ARIMA:", rmse_3_arima)
```
