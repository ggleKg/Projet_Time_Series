```python
## Question 1

import pandas as pd
import matplotlib.pyplot as plt

dpaes = pd.read_excel("/Users/baba/Desktop/dpae-mensuelles-france-entiere.xlsx",names= ['Anne' ,'Trimestre','Dernier jour du mois','Dure de contrat','Nature de contrat','DPAE (brut)','DPAE (cvs)'])

# Convertir la colonne 'Dernier jour du mois' en type datetime
dpaes['Dernier jour du mois'] = pd.to_datetime(dpaes['Dernier jour du mois'])

cdd = dpaes[dpaes["Nature de contrat"] == "CDD de plus d'un mois"]
cdi = dpaes[dpaes["Nature de contrat"] == "CDI"]

cdd = cdd.sort_values(by='Dernier jour du mois')
cdi = cdi.sort_values(by='Dernier jour du mois')




## Pour les CDD



# Tracer la variation de DPAE (brut) au fur et  mesure du temps pour les CDD
plt.figure(figsize=(12, 6))
plt.plot(cdd['Dernier jour du mois'], cdd['DPAE (brut)'])
plt.xlabel('Date')
plt.ylabel('DPAE (brut)')
plt.title('Variation de DPAE (brut) pour les CDD au fur et  mesure du temps')
plt.show()

plt.figure()

# Calculer la moyenne et l'cart type de DPAE (brut) par anne 
mean_by_year = cdd.groupby('Anne')['DPAE (brut)'].mean()
std_by_year = cdd.groupby('Anne')['DPAE (brut)'].std()
median_by_year = cdd.groupby('Anne')['DPAE (brut)'].median()

mean_by_year.plot()
std_by_year.plot()
median_by_year.plot()
plt.legend(['Mean', 'Standard Deviation', 'Median'])
plt.title('Statistiques de DPAE (brut) pour les CDD au fur et  mesure des annes')
plt.show()


# Calculer la moyenne et l'cart type de DPAE (brut) par trimestre
mean_by_trimester = cdd.groupby(['Anne', 'Trimestre'])['DPAE (brut)'].mean()
std_by_trimester = cdd.groupby(['Anne', 'Trimestre'])['DPAE (brut)'].std()
median_by_trimester = cdd.groupby(['Anne','Trimestre'])['DPAE (brut)'].median()

plt.figure()
mean_by_trimester.plot()
std_by_trimester.plot()
median_by_trimester.plot()
plt.title('Statistiques de DPAE (brut) pour les CDD au fur et  mesure des annes')
plt.legend(['Mean', 'Standard Deviation', 'Median'])
plt.show()

# Afficher les botes  moustaches par anne
cdd.boxplot(column=['DPAE (brut)'], by='Anne', figsize=(10, 6))
plt.title("Botes  moustaches pour les CDD par anne")
plt.ylabel("DPAE (brut)")
plt.show()

# Afficher les botes  moustaches par trimestre
cdd.boxplot(column=['DPAE (brut)'], by='Trimestre', figsize=(10, 6))
plt.title("Botes  moustaches pour les CDD par trimestre")
plt.ylabel("DPAE (brut)")
plt.show()


```

```python
# Question 4

from sklearn.metrics import mean_squared_error
# Calculer les prdictions
forecast_cdd = fit1.forecast(len(test_cdd))
forecast_cdi = fit2.forecast(len(test_cdi))

forecast_1 = fit1.forecast(len(test_1))
forecast_2 = fit2.forecast(len(test_2))
forecast_3 = fit3.forecast(len(test_3))



# Concatner les valeurs relles et les valeurs prdites
y_true_cdd = test_cdd
y_pred_cdd = forecast_cdd

y_true_cdi = test_cdi
y_pred_cdi = forecast_cdi

y_true_1 = test_1
y_pred_1 = forecast_1

y_true_2 = test_2
y_pred_2 = forecast_2

y_true_3 = test_3
y_pred_3 = forecast_3


# Calculer l'erreur MSE
rmse_cdd = mean_squared_error(y_true_cdd, y_pred_cdd, squared=False)
rmse_cdi = mean_squared_error(y_true_cdi, y_pred_cdi, squared=False)
rmse_1 = mean_squared_error(y_true_1, y_pred_1, squared=False)
rmse_2 = mean_squared_error(y_true_2, y_pred_2, squared=False)
rmse_3 = mean_squared_error(y_true_3, y_pred_3, squared=False)


# Afficher l'erreur MSE
print("Rooted Mean Squared Error (RMSE) CDD: ", rmse_cdd)
print("Rooted Mean Squared Error (RMSE) CDI: ", rmse_cdi)
print("Rooted Mean Squared Error (RMSE) secteur1: ", rmse_1)
print("Rooted Mean Squared Error (RMSE) secteur2: ", rmse_2)
print("Rooted Mean Squared Error (RMSE) secteur3: ", rmse_3)


```

```python
## Question 5
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm

# Entranement du modle ARIMA pour DPAE CDD
model_dpaecdd = ARIMA(train_cdd, order=(2, 0, 1)).fit()

# Entranement du modle ARIMA pour DPAE CDI
model_dpaecdi = ARIMA(train_cdi, order=(2, 0, 1)).fit()

# Entranement du modle ARIMA pour EESSA (secteur slectionn)
model_eessa = sm.tsa.ARIMA(train_c1, order=(2,0, 1)).fit()
model_eessa2 = sm.tsa.ARIMA(train_c2, order=(2,0, 1)).fit()
model_eessa3 = sm.tsa.ARIMA(train_c3, order=(2,0, 1)).fit()



# Prvisions pour DPAE CDD
pred_dpaecdd = model_dpaecdd.predict(start=len(train_cdd), end=len(train_cdd)+len(test_cdd)-1)

# Prvisions pour DPAE CDI
pred_dpaecdi = model_dpaecdi.predict(start=len(train_cdi), end=len(train_cdi)+len(test_cdi)-1)

# Prvisions pour EESSA (secteur slectionn)
pred_eessa = model_eessa.predict(start=len(train_c1), end=len(train_c1)+len(test_1)-1)
pred_eessa2 = model_eessa2.predict(start=len(train_c2), end=len(train_c2)+len(test_2)-1)
pred_eessa3 = model_eessa3.predict(start=len(train_c3), end=len(train_c3)+len(test_3)-1)


from sklearn.metrics import mean_absolute_error, mean_squared_error
from scipy.stats import pearsonr

# valuation des prvisions pour DPAE CDD
mae_cdd_arima = mean_absolute_error(test_cdd, pred_dpaecdd)
rmse_cdd_arima = mean_squared_error(test_cdd, pred_dpaecdd, squared=False)
print("RMSE pour les donnes DPAE (CDD) avec ARIMA:", rmse_cdd_arima)
# valuation des prvisions pour DPAE CDI
mae_cdi_arima = mean_absolute_error(test_cdi, pred_dpaecdd)
rmse_cdi_arima = mean_squared_error(test_cdi, pred_dpaecdd, squared=False)
print("RMSE pour les donnes DPAE (CDI) avec ARIMA:", rmse_cdi_arima)

# valuation des prvisions pour Secteur1
mae_1_arima = mean_absolute_error(test_1, pred_eessa)
rmse_1_arima = mean_squared_error(test_1, pred_eessa, squared=False)
print("RMSE pour les donnes EESSA avec ARIMA:", rmse_1_arima)

# valuation des prvisions pour Secteur2
mae_2_arima = mean_absolute_error(test_2, pred_eessa2)
rmse_2_arima = mean_squared_error(test_2, pred_eessa2, squared=False)
print("RMSE pour les donnes EESSA avec ARIMA:", rmse_2_arima)

# valuation des prvisions pour Secteur3
mae_3_arima = mean_absolute_error(test_3, pred_eessa3)
rmse_3_arima = mean_squared_error(test_3, pred_eessa3, squared=False)
print("RMSE pour les donnes EESSA avec ARIMA:", rmse_3_arima)
```
